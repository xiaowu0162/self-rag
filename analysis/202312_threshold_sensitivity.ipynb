```python
def plot_threshold_accuracy_tradeoff(task, threshold, accuracy, frequency):
    # Creating two subplots sharing the same x-axis
    fig, ax1 = plt.subplots(figsize=(3,2))
    
    # Plotting the accuracy vs threshold (x-axis) on the left y-axis
    color = 'teal'
    ax1.set_xlabel('Threshold')
    ax1.set_ylabel('Accuracy', color=color)
    ax1.plot(threshold, accuracy, color=color, marker='o')
    ax1.tick_params(axis='y', labelcolor=color)
    
    # Creating another y-axis for the frequency shared with ax1
    ax2 = ax1.twinx()
    color = 'blueviolet'
    ax2.set_ylabel('RAG Frequency', color=color)
    ax2.plot(threshold, frequency, color=color, marker='o')
    ax2.tick_params(axis='y', labelcolor=color)
    
    # Setting the title
    plt.title(task)
    
    # Adding spines (hard lines) for all four sides of the plot
    ax1.spines['top'].set_visible(True)
    ax1.spines['right'].set_visible(True)
    ax1.spines['bottom'].set_visible(True)
    ax1.spines['left'].set_visible(True)
    
    # Showing the plot
    plt.show()
```
